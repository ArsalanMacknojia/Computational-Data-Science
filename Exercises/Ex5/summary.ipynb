```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Read data from .csv into dataFrames
filenames = ["data-1.csv", "data-2.csv", "data-3.csv", "data-4.csv", "data-5.csv", "data-6.csv"]
data_files = {}

for filename in filenames:
    data_files[filename] = pd.read_csv(filename)
```

```python
# Calculate mean, standard deviation, range (min and max) and correlation coefficient between x and y variables
for file in data_files:
    df = data_files[file]
    mean_x = df['x'].mean()
    mean_y = df['y'].mean()
    
    std_x = df['x'].std()
    std_y = df['y'].std()
    
    min_x = df['x'].min()
    min_y = df['y'].min()
    max_x = df['x'].max()
    max_y = df['y'].max()
    
    corr_coeff = df.corr(method ='pearson')
    
    print("------------------------------------------------------------------------")
    print("Data Set: {}".format(file))
    print("Mean X: {}, Mean Y: {}".format(mean_x, mean_y))
    print("Standard Deviation X: {}, Standard Deviation Y: {}".format(std_x, std_y))
    print("Min X: {}, Min Y: {}".format(min_x, min_y))
    print("Max X: {}, Max Y: {}".format(max_x, max_y))
    print("Correlation Coefficient between X and Y:\n {}".format(corr_coeff))
```

```python
# Scatter plot of x vs y
for file in data_files:
    df = data_files[file]
    plt.figure(figsize=(20,10))
    plt.scatter(df['x'], df['y'])
    plt.title(file)
    plt.xlabel("x-values")
    plt.ylabel("y-values")
    plt.show
```
